```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
df.head()
```

```python
# target variable distribution
sns.countplot(df['Diabetes_012'].astype(str), order = df['Diabetes_012'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('Diabetes Class')
plt.title('Diabetes Class Distribution')
plt.show()
```

In the above chart of our target variable we can see this dataset is unbalanced with no diabetes making of a large majority of survey responses. 

```python
# histogram of bmi grouped by Diabetes_012
sns.histplot(data=df, x='BMI', hue='Diabetes_012', bins=50)
plt.xlabel('BMI')
plt.ylabel('Count')
plt.title('BMI Distribution')
plt.show()
```

In the above chart we can see the distribution of BMI among survey respondents is right skewed between 20-40 BMI.

```python
# histogram of general health
sns.histplot(data=df, x='GenHlth', hue='Diabetes_012', discrete=True, multiple='dodge')
plt.xlabel('General Health')
plt.ylabel('Count')
plt.title('General Health Distribution')
plt.show()
```

The above chart shows a majority of the non diabetic class fall in the 1-3 general health level

```python
# histogram of Age
sns.histplot(data=df, x='Age', hue='Diabetes_012', discrete=True, multiple='dodge')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')
plt.show()
```

The above chart shows a slight left-skew but there does not appear to be any relationship between age and class
